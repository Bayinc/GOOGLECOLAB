## 1.1 Carga de datos almacenados en Colab

- Estos datos vienen por defecto cargados en Colab, y los podemos utilizar como queramos.
- Es clave reconocer la ruta donde estos datos están almacenados.
- La ruta del archivo la podemos obtener con clic derecho sobre el archivo.

```python
# Carga la librería "pandas" bajo el nombre de "pd"
import pandas as pd
```

```python
# Se lee: En "datos" guardamos el archivo .csv leído con la función "read_csv" de la librería "pd"
# "datos" será un objeto de tipo "dataframe" (de pandas)
# Dataframe es una estructura de datos, así como las listas

datos = pd.read_csv("/content/sample_data/california_housing_test.csv")
```

```python
# head() es un método de los objetos "dataframe" de pandas que muestra las primeras filas de una data.
# Un "método" es prácticamente una función, solo que se puede aplicar sobre ciertos objetos.

datos.head()
```

```python
# Veamos las 10 primeras filas de la data
# Como head() es un método (función), podemos indicar valores dentro

datos.head(10)
```

```python
# columns es un atributo de un dataframe, no lleva (), siempre retorna lo mismo

datos.columns
```

### 1.2.2 CSV
